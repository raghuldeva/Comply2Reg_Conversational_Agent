```python
from langchain_groq import ChatGroq

groq_api_key = "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"
model = ChatGroq(groq_api_key=groq_api_key, model_name="Llama3-8b-8192", temperature=0)
```

```python
from langchain.prompts import PromptTemplate

# prompt = PromptTemplate(
#     template="""<|begin_of_text|><|start_header_id|>system<|end_header_id|> You are an expert in summarising a document that should say what the document is about . Document content: {content}. <|eot_id|><|start_header_id|>assistant<|end_header_id|>""",
#     input_variables=["content"],
# )


prompt = PromptTemplate(
    template="""<|begin_of_text|><|start_header_id|>system<|end_header_id|> You are an expert in summarising a document that should say what the document is about. Document content: {content}. Give a summary in very short about 3 lines in a single paragraph<|eot_id|><|start_header_id|>assistant<|end_header_id|>""",
    input_variables=["content"],
)
```

```python
from langchain_core.output_parsers import StrOutputParser
prompt_question = prompt | model | StrOutputParser()
```

```python
content = pages[1].page_content
print(prompt_question.invoke({'content':content}))
```

```python
all_page_summaries = ""
content_list = []
for i in range(0, len(pages)):
    content = pages[i].page_content
    content_list.append(pages[i])
    print(content)
    print(i)
    all_page_summaries += prompt_question.invoke({'content':content})
    print(all_page_summaries)
    print("*******************************************************")
```
